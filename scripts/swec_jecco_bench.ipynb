```julia
# interpolate swec and compare exactly on the same u points swec and jecco
swec_ug1 = (swec_Bg1[1,1:12,1]);
swec_z   = (swec_Bg1[:,1,2]);
bg1spl2=Spline2D(swec_ug1, swec_z, 1.5*bg1[:,:]);
```

```julia
# compare swc and jecco at the same u point using bg1spl2
itp_bg1 = zeros(length(swec_ug1),length(swec_z));
for i in 1:12
    for j in 1:128
        u = ujecco[i]
        z = xjecco[j]
        itp_bg1[i,j] = bg1spl2(u,z)
    end
end
```

```julia
# careful there is extrapolation for i=12, since ujecco[12]-swec_ug1[12] = 0.00170370868554659
i = 11
itp_b1reldif = (itp_bg1[i,:] - b1in[i,:])./itp_bg1[i,:];
print("itp_B1_in max_rel_error = ", maximum(abs.(itp_b1reldif))*100," %")
```

```julia
# out domain
B1_out, info_out=get_field(ts, it=2000, field="B1 c=2");
u_out_jecco=info_out.coords[1];
x_out_jecco=info_out.coords[2];
#heatmap(B1_in[3,:,:], aspect_ratio=:equal)
plot(B1_out[56,:,1])
```

```julia
# interpolate swec fot g2 and compare exactly on the same u points swec and jecco
bg2       = transpose(swec_Bg2[:,:,3])
swec_ug2  = (swec_Bg2[1,:,1]);
swec_zg2 = (swec_Bg2[:,1,2]);
bg2spl2   = Spline2D(swec_ug2, swec_zg2, 1.5*bg2[:,:]);
```

```julia
surface(1.5*bg2)
```

```julia
b1out = B1_out[:,:,1]
surface(b1out[:,:])
```

```julia
# compare swc and jecco at the same u point using bg2spl2
itp_bg2 = zeros(length(u_out_jecco),length(x_out_jecco));
for i in 1:length(u_out_jecco)
    for j in 1:length(x_out_jecco)
        u = u_out_jecco[i]
        z = x_out_jecco[j]
        itp_bg2[i,j] = bg2spl2(u,z)
    end
end
```

```julia
# comparison for out domain; swec;s out domain goes to lower u that jecco's. the upper limit is the same.
i = 56
itp_b2reldif = (itp_bg2[i,:] - b1out[i,:])./itp_bg2[i,:];
print("itp_B1_out max_rel_error = ", maximum(abs.(itp_b2reldif))*100," %")
```
